```python
base_path = "/content/drive/MyDrive"
with open(f"{base_path}/task5_text_file.txt", "w+") as file:
  file.write("Первая строка текстового файла\n")
  file.write("Ещё одна строка")

with open(f"{base_path}/task5_text_file.txt", "r") as file:
  print(f"Содержимое записанного файла: \n{file.read()}")
```

```python
students = [
    ["Иванов И.И.", 20, "ИВТ-101"],
    ["Петров П.П.", 22, "ИВТ-102"],
    ["Сидорова А.А.", 21, "ИВТ-103"]
]

with open(f"{base_path}/students.csv", "w") as file:
  for student in students:
    file.write(",".join(map(str, student)) + "\n")

print("Файл students.csv успешно сохранен в Google Drive")
```

```python
import os

file_to_process_path = f"{base_path}/task5_text_file.txt"
processed_file_path = f"{base_path}/task5_text_file_processed.txt"

with open(file_to_process_path, "r") as file:
  content = file.read()
print(f"Содержимое оригинального файла '{os.path.basename(file_to_process_path)}':\n{content}")

with open(processed_file_path, "w") as file:
  modified_content = content + "\nДобавлена новая строка во время обработки.\nЕщё одна новая строка."
  print(f"\nОбработанное содержимое:\n{modified_content}")
  file.write(modified_content)

print(f"\nФайл '{os.path.basename(processed_file_path)}' успешно сохранен в Google Drive с обработанным содержимым.")

with open(processed_file_path, "r") as file:
    print(f"\nСодержимое сохраненного файла '{os.path.basename(processed_file_path)}':\n{file.read()}")
```
