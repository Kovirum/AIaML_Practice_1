```python
import os

# Создание и запись в файл
with open("task2_example.txt", "w") as f:
    f.write("Первая строка\n")
    f.write("Вторая строка\n")

# Чтение файла
with open("task2_example.txt", "r") as f:
    content = f.read()
    print("Содержимое файла:\n", content)

# Проверка существования файла
print("Файл существует:", os.path.exists("task2_example.txt"))
```

```python
# Циклическое перезапись данных в файл по строкам
result = ""
while True:
    content = input("Введите строку текста или [break] для остановки и записи результата в файл")
    if content == "[break]":
        break
    else:
        result += f"{content}\n"
    
with open("task2_example.txt", "w") as f:
    f.write(result)
    print("Содержимое файла обновлено!")
```

```python
# Удаление файла
os.remove("task2_example.txt")
print("Файл удален.")
```
